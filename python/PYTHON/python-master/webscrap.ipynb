```python
import urllib2
import pandas as pd
from bs4 import BeautifulSoup as bs
source = urllib2.urlopen("http://apps.fas.usda.gov/export-sales/wkHistData.htm")
data = source.read()
soup = bs(data)
base_url = s[:s.rfind('/')]+"/"
for a in soup.findAll('a'):
    url = base_url + a.attrs.get('href','')
    data = pd.read_html(url)
    print data
```
